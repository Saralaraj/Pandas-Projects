```python
df.groupby('job')['day'].agg('sum').to_frame()
```

```python
df.groupby('job')['age'].agg(['min','max']).sort_values(by = 'job')
```

```python
df.groupby('job')['campaign'].agg('count').to_frame()
```

```python
df.groupby(['job','education']).agg({'campaign':'count',
                                    'age':['min','max']}).sort_values(by = 'education')
```

```python
df.groupby('job')['balance'].agg('mean').to_frame()
```

```python
df.groupby('Target').agg({'month':['min','max']})
```

```python
df.groupby(['job','age']).agg({'day':'sum',
                              'Target':'count',
                              'balance':'mean',
                              'campaign':['min','max']}).sort_values('job')
```
