```python
import pandas as pd
import numpy as np
df =pd.read_csv('Mall_customers.csv')
df
```

```python
df = df.rename(columns = {'Genre':'Gender','Annual_Income_(k$)' : 'Income'}) 
df 
```

```python
start_date = np.datetime64('2023-01-01')
end_date = np.datetime64('2023-12-31')
df['Date'] = np.random.choice(pd.date_range(start = start_date,
                                           end = end_date),size = len(df))
df
```

```python
num_orders = len(df)
order_start = 1000
order_end = 9999
df['order_id'] = np.random.randint(order_start,order_end,size =num_orders)
df
```

```python
rfm = df.groupby('CustomerID').agg({
    'Date' : lambda x : (df['Date'].max() - x.max()).days,
    'order_id' : 'count',
    'Spending Score (1-100)': 'sum'
    
}).reset_index()
rfm
```

```python
rfm = df.groupby('CustomerID').agg({

    'Date' : lambda x: (df['Date'].max() - x.max()).days,
    'order_id' : 'count',
    'Spending Score (1-100)' : 'sum'
}).reset_index()
rfm
```

```python
rfm.columns = ['CustomerID','Recency','Frequency','Monetary']
rfm
```

```python
rfm['R_rank'] = pd.qcut(rfm['Recency'],4,['4','3','2','1'])
rfm['F_rank'] = pd.qcut(rfm['Frequency'].rank(method = 'first'),4,['1','2','3',4])
rfm['M_rank'] = pd.qcut(rfm['Monetary'].rank(method = 'first'),4,['1','2','3',4])
rfm
rfm['RFM_Score'] =  rfm['R_rank'].astype(str)+rfm['F_rank'].astype(str)+rfm['M_rank'].astype(str)
rfm
```
