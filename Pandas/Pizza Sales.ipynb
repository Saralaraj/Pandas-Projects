```python
#16
def categorize_price (price):
    if price < 10:
        return 'low'
    elif 10 <= price <=15:
        return 'medium'
    else:
        return 'high'
df['price_category'] = df['price'].apply(categorize_price)
sales_summary = df.groupby(['pizza_type','price_category'])['price'].agg('sum')

print(sales_summary)

```
