```python
# 14 How can you calculate the percentage contribution of each quarter to the total sales for each brand
df['Total Sales'] = df.groupby('Brand')['Q1 2019'].transform('sum')
df['Q1 percentage contribution 2019'] = df['Q1 2019']/df['Total Sales']*100
df['Q2 percentage contribution 2019'] = df['Q2 2019']/df['Total Sales']*100
df['Q3 percentage contribution 2019'] = df['Q3 2019']/df['Total Sales']*100
df['Q4 percentage contribution 2019'] = df['Q4 2019']/df['Total Sales']*100
df
df['Total Sales'] = df.groupby('Brand')['Q1 2020'].transform('sum')
df['Q1 percentage contribution 2020'] = df['Q1 2020']/df['Total Sales']*100
df['Q2 percentage contribution 2020'] = df['Q2 2020']/df['Total Sales']*100
df['Q3 percentage contribution 2020'] = df['Q3 2020']/df['Total Sales']*100
df['Q4 percentage contribution 2020'] = df['Q4 2020']/df['Total Sales']*100
df
```

**Interpretation**:

      - In this above code,we can see the percentage contribution of each quarter to the total sales for each brand

```python
# 16 How can you rename the columns Q1 Sales, Q2 Sales, Q3 Sales, and Q4 Sales to Quarter 1, Quarter 2, Quarter 3, and Quarter 4
df.rename(columns = {'Q1 2019':'Quarter 1 2019','Q2 2019':'Quarter 2 2019','Q3 2019':'Quarter 3 2019','Q4 2019':'Quarter 4 2019',
                     'Q1 2020':'Quarter 1 2020','Q2 2020':'Quarter 2 2020','Q3 2020':'Quarter 3 2020','Q4 2020':'Quarter 4 2020'},inplace = True)
print(df)
```

**Interpretation**:

```python
# 17 How can you fill any missing sales data with the average sales of that quarter
average_sales = df.groupby('Brand')['Q1 2019'].agg('mean')
df['Sales'] = df['Q1 2019'].fillna(average_sales)
df.drop(columns = 'Q1 2019',inplace = True)
print(df)
```

**Interpretation**

      - In this above code, we can see the missing sales data with the average sales of that quarter

```python
# 18 How can you calculate the year-over-year sales growth for each brand
df[2019] = df['Q1 2019']+df['Q2 2019']+df['Q3 2019']+df['Q4 2019']
df[2020] = df['Q1 2020']+df['Q2 2020']+df['Q3 2020']+df['Q4 2020']
df['yoy growth percentage'] = df.groupby('Brand')['Q1 2019'].pct_change()*100
print(df)
```

**Interpretation**

            In this above code, we can see the year-over-year sales growth for each brand

```python
# 19 How can you filter the DataFrame to show only the top 5 brands by total sales
a = df.groupby('Brand')['Q1 2019'].agg('sum')
b = a.sort_values(ascending = False).to_frame()
top_5_brand = b.head(5)
print(top_5_brand)
```

**Interpretation**
