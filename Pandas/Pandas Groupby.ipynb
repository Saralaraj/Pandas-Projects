```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'PRODUCT' : ['A','A','A','B','B','B','C','C','C','C'], 
                   'REGION' : ['North','South','East','North','South','East','North','South','East','East'],
                   'SALES' : [100,150,200,100,200,300,400,500,600,700],
                   'PROFIT' : [20,30,40,10,20,30,50,60,70,80],
                   'MONTH' : ['Jan','Feb','Mar','Jan','Feb','Mar','Jan','Feb','Mar','Apr']
})
df
```

```python
#1) Total Sales by Product
df.groupby(['PRODUCT'])['SALES'].agg('sum').to_frame()
```

```python
# 2) Average profit by Region
df.groupby(['REGION'])['PROFIT'].agg('mean').to_frame()
```

```python
# 3) Total sales by Region

df.groupby(['REGION'])['SALES'].agg('sum').to_frame()
```

```python
# 4) Total sales and Profit by Product

df.groupby('PRODUCT').agg({'PROFIT' : 'sum',
                                 'SALES' : 'sum'})
```

```python
# 5) Average Sales by Month

df.groupby(['MONTH'])['SALES'].agg('mean').to_frame()
```

```python
# 6) Total Sales by product and Region

df.groupby(['REGION','PRODUCT'])['SALES'].agg('sum').to_frame()
```

```python
# 7) Maximum Sales by Region 

df.groupby('REGION').agg({'SALES' : ['max']})
```

```python
# 8) Minimun Profit by Product

df.groupby('PRODUCT').agg({'PROFIT' : ['min']})
```

```python
# 9) count of Sales Entries by product

df.groupby('PRODUCT')['SALES'].agg('count').to_frame()
```

```python
# 10) Count of sales entries by Region

df.groupby('REGION')['SALES'].agg('count').to_frame()
```

```python
# 11) Total Profit by product and Month

df.groupby(['PRODUCT','MONTH'])['PROFIT'].agg('sum').to_frame()
```

```python
# 12) Mean and standard Deviation of sales by Product

df.groupby('PRODUCT')['SALES'].agg(['mean','std'])
```

```python
# 13) Sum of Sales and Profit by Product and Region

df.groupby(['PRODUCT','REGION']).agg({'SALES' :'sum','PROFIT' : 'sum' })
```

```python
#14) Total Sales by Month and region

df.groupby(['MONTH','REGION'])['SALES'].agg('sum').to_frame()
```

```python
# 15) Average Sales and Profit by Region

df.groupby(['REGION']).agg({'SALES' : 'mean','PROFIT' : 'mean'})
```

```python
# 16) Median Sales by Product

df.groupby('PRODUCT')['SALES'].agg('median').to_frame()
```

```python
#17) Average Sales by product and month

df.groupby(['PRODUCT','MONTH']).agg({'SALES': 'mean'})
```

```python
# 18) Sum of Sales,Profit by Region Where Sales >300

a=df[df['SALES']>300]
b = a.groupby('REGION').agg({'SALES' : 'sum','PROFIT' : 'sum'})
print(b)
```

```python
# 19) Total Sales by Product sorted by Sales

df.groupby(['PRODUCT','SALES']).agg({'SALES': 'sum'} )
```

```python
# 20) Total Profit by Product with Sales Greater than 100
import pandas as pd

a = df[df['SALES']>100]
b = a.groupby('PRODUCT')['PROFIT'].agg('sum').to_frame()
print(b) 
```
